```python
import numpy as np 
import pandas as pd 
import seaborn as sns
from matplotlib import pyplot as plt
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
def missing_value_table(df):
    missing_value = df.isnull().sum()
    missing_value_percent = 100 * df.isnull().sum() / len(df)
    missing_value_table = pd.concat([missing_value,missing_value_percent], axis = 1)
    missing_value_table_end = missing_value_table.rename(columns = {0 : 'Missing Value', 1 : '% deÄŸeri'})
    return missing_value_table_end
```

```python
missing_value_table(df)
```

```python
df[df["Embarked"].isnull()]
```

```python
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
df["Embarked"] = imp_mean.fit_transform(df[["Embarked"]])
```

```python
df[df["Age"].isnull()]
```

```python
from sklearn.impute import KNNImputer
df_numeric = df.select_dtypes(include = ["float64","int64"])
var_names = df_numeric.columns
n_df = np.array(df_numeric)
```

```python
n_df[0:10]
```

```python
imputer = KNNImputer(n_neighbors=4)
new_data = imputer.fit_transform(n_df)
df2=pd.DataFrame(new_data, columns=var_names)
df[["Age"]] = df2.Age
```

```python
df[df["Cabin"].isnull()]
```

```python
df.drop(["Cabin"], axis =1, inplace = True)
```

```python
df.drop(["Name","PassengerId","Ticket"], axis =1, inplace = True)
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
lbe = LabelEncoder()
df["Sex"] = lbe.fit_transform(df["Sex"])
```

```python
df = pd.concat([df, pd.get_dummies(df["Embarked"])], axis=1).drop("Embarked", axis = 1)
df
```

```python
y=df["Survived"]
X=df[["Sex", "Pclass", "Age","SibSp","Parch","Fare","C","Q","S"]]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.20, random_state=111)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(solver='lbfgs', multi_class="ovr")
```

```python
log_reg.fit(X_train, y_train)
```

```python
train_accuracy = log_reg.score(X_train, y_train)
test_accuracy = log_reg.score(X_test, y_test)

print('One-vs.-Rest', '-'*30, 
      'Accuracy on Train Data : {:.2f}'.format(train_accuracy), 
      'Accuracy on Test Data  : {:.2f}'.format(test_accuracy), sep='\n')
```

```python
log_reg_mnm = LogisticRegression(multi_class='multinomial', solver='lbfgs')
log_reg_mnm.fit(X_train, y_train)

train_accuracy = log_reg_mnm.score(X_train, y_train)
test_accuracy = log_reg_mnm.score(X_test, y_test)

print('Multinomial (Softmax)', '-'*20, 
      'Accuracy on Train Data : {:.2f}'.format(train_accuracy), 
      'Accuracy on Test Data  : {:.2f}'.format(test_accuracy), sep='\n')
```

```python
print(log_reg.predict(X)[0:10])
```

```python
C_values = [0.001,0.01, 0.1,1,10,100, 1000]

accuracy_values = pd.DataFrame(columns=['C_values', 'Train Accuracy', 'Test Accuracy'])

for c in C_values:
    # Apply logistic regression model to training data
    lr = LogisticRegression(penalty = 'l2', C = c, random_state = 0, solver='lbfgs', multi_class='multinomial')
    lr.fit(X_train, y_train)
    accuracy_values = accuracy_values.append({'C_values': c,
                                              'Train Accuracy': lr.score(X_train, y_train),
                                              'Test Accuracy': lr.score(X_test, y_test)
                                             }, ignore_index=True)
display(accuracy_values) 
```
