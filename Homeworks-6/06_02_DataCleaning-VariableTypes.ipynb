```python
import pandas as pd
import numpy as np
movie = pd.read_csv("https://bootrain-lms-assets.s3.eu-central-1.amazonaws.com/bootrain-lms-static/datasets/movie.csv")
```

```python
movie
```

```python
movie.title_year.unique() # The color string takes a value. In this case, we can say it is categorical and nominal.
```

```python
def make_categorical(title_year):
    if title_year >= 2000 :
        return "21.yy"
    elif title_year < 2000 :
        return "20.yy"
```

```python
movie['year_grup'] = movie['title_year'].apply(make_categorical)
display(movie.head())
movie.groupby("year_grup")["title_year"].count()
```
