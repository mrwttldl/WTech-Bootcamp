```python
list = ["ENROLL", "TOTAL_REVENUE", "FEDERAL_REVENUE", 
             "STATE_REVENUE", "LOCAL_REVENUE", "TOTAL_EXPENDITURE", 
             "INSTRUCTION_EXPENDITURE", "SUPPORT_SERVICES_EXPENDITURE",
             "OTHER_EXPENDITURE", "CAPITAL_OUTLAY_EXPENDITURE", "GRADES_PK_G", 
             "GRADES_KG_G", "GRADES_4_G", "GRADES_8_G", "GRADES_12_G", "GRADES_1_8_G", 
             "GRADES_9_12_G", "GRADES_ALL_G"]

for col in list:
    df2.loc[:, col].fillna(df2.loc[:, col].mean(), inplace=True)
```

3.Now it's time to consider the factor of time! Review your answer in question 2 and fill in the missing data based on the values observed during that year. For example, if you want to fill a value with an average value, calculate the average of that year.

```python
df3 = df.copy()
```

```python
years = df3["YEAR"].unique()

for col in list:
    for year in years:
        YIL_SART = df3["YEAR"] == year
        df3.loc[YIL_SART, col] = df3.loc[YIL_SART, col].fillna(df3[YIL_SART][col].mean())
```

```python
import missingno
from matplotlib import pyplot as plt
missingno.matrix(df3, figsize = (15,10))
plt.title("Missing Values (df)")
plt.show()
```

```python
df3.isnull().sum()/df3.isnull().count()
```
