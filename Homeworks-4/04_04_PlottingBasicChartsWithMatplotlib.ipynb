In this assignment, you will continue work with the [Coronavirus Source Data](https://ourworldindata.org/coronavirus-source-data). You will plot different chart types. Don't forget to set titles and axis labels.

```python
import pandas as pd

import matplotlib.pyplot as plt

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
```

1. Plot a bar chart for total cases of the 20 countries that havebiggest numbers.

```python
df2 =pd.DataFrame(df.groupby("location")['total_cases'].sum().sort_values(ascending=False).head(20))
df2
```

```python
plt.figure(figsize=(20, 10))
plt.title('20 countries with the highest total number of cases')
plt.bar(df2.index,df2["total_cases"], color = "#22B573")
plt.xlabel('Countries')
plt.ylabel('Total Cases')
plt.xticks(rotation = 90, fontsize = 9)
plt.show()
```

2. Plot a histogram for daily deaths for any country you choose. Make three subplots for different bins.

```python
us = df[df["location"] == "United States"]
us
```

```python
import warnings
warnings.filterwarnings('ignore')
title_style_small = {'family': 'Century Gothic', 'color': 'darkred', 'size': 10 }
axis_style_small  = {'family': 'Century Gothic', 'color': 'darkblue', 'size': 7 }

plt.figure(figsize=(15, 9))

plt.subplot(2,2,1)
plt.title('Total Cases of United States', fontdict = title_style_small)
plt.hist(us['new_deaths'], color = "blue", bins =10)
plt.xlabel("New Deaths")
plt.ylabel("Counts")

plt.subplot(2,2,2)
plt.title('Total Cases of United States', fontdict = title_style_small)
plt.hist(us['new_deaths'], color = "red", bins =20)
plt.xlabel("New Deaths")
plt.ylabel("Counts")

plt.subplot(2,2,3)
plt.title('Total Cases of United States', fontdict = title_style_small)
plt.hist(us['new_deaths'], color = "green", bins =30)
plt.xlabel("New Deaths")
plt.ylabel("Counts")

plt.tight_layout()
plt.show()
```

3. Plot a scatter plot of new cases and new death for Germany and France.

```python
plt.figure(figsize=(15, 8))
plt.title('New Cases & New Deaths')
for status in ['Germany', 'France']:
    plt.scatter(df[df.location == status].new_cases, 
                df[df.location == status].new_deaths, label = status)
plt.xlabel('New Cases')
plt.ylabel('New Deaths')
plt.legend()
plt.show()
```

4. Plot a boxplot for daily deaths for any country you choose.

```python
turkey=df[(df['location']=='Turkey')]
turkey
```

```python
plt.figure(figsize=(8, 6))
plt.title('Deaths daily chart for Turkey')
plt.boxplot(turkey['new_deaths'].dropna())
plt.xlabel('Turkey')
plt.ylabel('Counts of Daily Deaths')
plt.show()
```

5. Calculate the total case for each continent and plot a pie chart 

```python
df3 =pd.DataFrame(df.groupby(by="continent")['total_cases'].sum())
df3
```

```python
plt.figure(figsize=(15, 8))
plt.title('Total case for each continent')
plt.pie(df3.total_cases,labels=df3.index,shadow=True, autopct='%1.1f%%',startangle=90)
plt.show()
```
