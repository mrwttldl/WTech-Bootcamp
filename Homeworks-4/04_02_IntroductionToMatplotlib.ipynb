1. Create a blank figure with matplotlib functions. Plot cosine function for the numbers beetwen -3$\pi$ and 3$\pi$ with 0.1 intervals.

```python
import matplotlib.pyplot as plt
import numpy as np
import math
in_array = np.arange(3*(-math.pi),3*(math.pi), 0.1)
out_array = np.cos(in_array) 
  
plt.plot(in_array, out_array, color = 'green') 
plt.title("numpy.cos()") 
plt.xlabel("X") 
plt.ylabel("Y") 
plt.show() 
```

2. This time, create the same plot in red color with object methods. 

```python
plt.plot(in_array, out_array, color = 'red') 
plt.title("numpy.cos()") 
plt.xlabel("X") 
plt.ylabel("Y") 
plt.show() 
```
